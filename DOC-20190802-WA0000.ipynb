```python
def dob():   # a function which asks for the dob and prints out in  a proper format
    m = input("Month you were born: ")
    d = input("Date you were born: ")
    y = input("Year you were born: ")
    
    print('mm/dd/yyyy: ', m, '/', d, '/', y)
    
    return m, d, y
```
